```python
import json
import pandas as pd
```

```python
# model_name = 'Llama3Point2Vision90B'
model_name = 'Gemini1Point5Pro'

queries_file = f'../../data/synthTRIPS/generated-queries/{model_name}_generated_queries.json'

with open(queries_file, 'r') as f:
    queries = json.load(f)
len(queries)
```

```python
queries[0]
```

```python
data = []
for q in queries:
    query_v = q.get('query_v')
    cities = sorted(q.get('city', []))[:5]
    data.append({'config_id': q.get('config_id'), 'query_v': query_v, 'top5_cities': cities})

df = pd.DataFrame(data)
print(df.shape)
df.head()
```
